```

#to find hcf of 2 numbers
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
if(num2 == 0):
  print("HCF :",num1)
else:
    while(num2 != 0):
      rem = num1 % num2
      num1 = num2
      num2 = rem
      print("HCF :",num1)
```

```
#greatest digit present in given number
num = int(input("Enter a number: "))

max_digit = 0

for digit in num:
    if int(digit) > max_digit:
        max_digit = int(digit)

print("Greatest digit is:", max_digit)
```
